```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
df=pd.read_csv("/content/synthetic_real_estate.csv")
print(df.head())
print(100*"#")
print("ISNULL COUNT IN ATTRIBUTES")
print(df.isnull().sum())
```

```python
fig, axes = plt.subplots(1, 1, figsize=(12, 5))

# Stacked bar chart for Property_Type vs Location
property_location = df.groupby(['Location', 'Property_Type']).size().unstack()
property_location.plot(kind='bar', stacked=True, ax=axes)
axes.set_title('Distribution of Property Type by Location')
axes.set_xlabel('Location')
axes.set_ylabel('Count')
axes.legend(title='Property Type')
axes.tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 1, figsize=(6, 5))

property_location = df.groupby(['Location', 'Property_Type']).size().unstack()

property_location.plot(kind='bar', stacked=True, ax=axes)

axes.set_title('Distribution of Property Type by Location')   # Updated title
axes.set_xlabel('Location')
axes.set_ylabel('Count')
axes.legend(title='Property Type')
axes.tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 1, figsize=(6, 5))

# Segmented bar chart for Property_Type vs Location
property_location = df.groupby(['Location', 'Property_Type']).size().unstack()
property_location_prop = property_location.divide(property_location.sum(axis=1), axis=0)
property_location_prop.plot(kind='bar', stacked=True, ax=axes)
axes.set_title('Proportion of Property Type by Location')
axes.set_xlabel('Location')
axes.set_ylabel('Proportion')
axes.legend(title='Property Type')
axes.tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```

```python
from statsmodels.graphics.mosaicplot import mosaic

# Create mosaic plot for Location and Property_Type
plt.figure(figsize=(8, 6))
mosaic(df, ['Location', 'Property_Type'], title='Mosaic Plot of Location and Property Type')
plt.show()
```

```python
sns.regplot(x='Area_sqft', y='Price', data=df)
plt.title('Scatter plot of Area_sqft vs Price with Regression Line')
plt.xlabel('Area (sqft)')
plt.ylabel('Price')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Location', y='Price', data=df)
plt.title('Mean Price by Location')
plt.xlabel('Location')
plt.ylabel('Mean Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='Area_sqft', hue='Location', fill=True)
plt.title('Kernel Density Estimate of Area_sqft by Location')
plt.xlabel('Area (sqft)')
plt.ylabel('Density')
plt.show()
```

```python
fig, axes = plt.subplots(1, 1, figsize=(8, 6))
sns.boxplot(x='Location', y='Price', data=df, ax=axes)
axes.set_title('Box plot of Price by Location')
axes.set_xlabel('Location')
axes.set_ylabel('Price')
plt.show()
```

```python
fig, axes = plt.subplots(1, 1, figsize=(8, 6))
sns.violinplot(x='Location', y='Price', data=df, ax=axes)
axes.set_title('Violin plot of Price by Location')
axes.set_xlabel('Location')
axes.set_ylabel('Price')
plt.show()
```

```python
fig, axes = plt.subplots(1, 1, figsize=(10, 6))
sns.swarmplot(x='Location', y='Price', data=df, ax=axes)
axes.set_title('Beeswarm plot of Price by Location')
axes.set_xlabel('Location')
axes.set_ylabel('Price')
plt.show()
```
